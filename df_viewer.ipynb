```python
import pandas as pd
from IPython.display import HTML, display

# Load your DataFrame
df = pd.read_pickle("2009-2019_batch_24_close_strategy_1_fl-fv_64-8_training_NQ_OIL_GOLD_ES_backtest_NQ_OIL_GOLD_ES_calib_0p3_LR_0p004.pkl")
#df=df[(df.symbol=="NQ")&(df.batch_size==30)]
#df=df[["model_name",	"batch_size",	"close_strategy",	"Sharpe_up_seen",	"Sharpe_down_seen",	"Sharpe_down_unseen",	"Sharpe_both_ways_seen",	"Sharpe_both_ways_unseen",	"Sharpe_up_unseen","seen_years",	"symbol",	"Image_up",	"Image_down",	"Image_bothways",	"Image_rr_up",	"Image_rr_down",	"Image_rr_bothways"]]
def display_scrollable_df(df, height=500):
    styles = f"""
    <style>
      .scrollable-table-container {{
        overflow: auto;
        max-height: {height}px;
      }}
      .scrollable-table-container table {{
        border-collapse: collapse;
        width: 100%;
      }}
      .scrollable-table-container th {{
        position: sticky;
        top: 0;
        background: #f0f0f0;
        z-index: 2;
      }}
    </style>
    """
    html_table = df.to_html(escape=False)
    return HTML(styles + f'<div class="scrollable-table-container">{html_table}</div>')

# Display the DataFrame with sticky headers
display(display_scrollable_df(df.sort_values(by="Sharpe_up_unseen", ascending=False)))
#display(display_scrollable_df(df))#(df.sort_values(by="" ,ascending=False)))
```
